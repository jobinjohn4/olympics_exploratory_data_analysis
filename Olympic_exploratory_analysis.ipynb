# Olympic Exploratory Data Analysis

The Olympic dataset for exploratory data analysis encompasses comprehensive information on athletes, countries, sports, events, and historical performances across various Olympic Games editions from 1896 to 2016. It includes athlete details like names and ages, country information, sports, event specifics, and medal results. I have utilized this dataset to uncover patterns and trends through visualizations and statistical analyses. Exploring athlete demographics, performance trends, and country-specific achievements enhances our understanding of the dynamic history of the Olympic Games, providing valuable insights into the world of international sports competitions.

### Data Cleaning 

Removing irrelevant column.

```python
main_df=main_df.drop(columns=['Games'])
```

### Athlete's Performance Analysis.

Top 20 Highest number of Athlete's participation by Country. 

```python
count=main_df.Country.value_counts()
```

```python
count.head(20).plot(kind='bar',color='#c634eb',ylabel="No. of Athlete's")
plt.title("No. Of Athlete's Participation from Various Countries")
plt.show()
```

Athlete's representation by Gender.

```python
sns.set_palette("icefire")
sns.countplot(main_df,x="Sex`
plt.ylabel("No. of Athlete's")
plt.title("Athlete's Distribution by Gender")
plt.show()
```

Countrywise Male and Female representation.

```python
country_gender=main_df.groupby(['Country','Sex'])['ID'].count()
country_gender.head(10).to_frame()
```

Number of Athlete's Distribution by Age.

```python
sns.histplot(main_df,x="Age",bins=20,kde=True,color='#a7f542')
plt.title("Athlete's Distribution by Age")
plt.show()
```

Average age of athletes participated each year.

```python
avg_age_year=main_df.groupby("Year")["Age"].mean()
avg_age_year.head(10)
```

```python
avg_age_year.plot(kind='bar',color='#c634eb',ylabel="Average Age of Athlete's")
plt.title("Average age of Athlete's by Years ")
plt.xticks(rotation=65)
plt.show()
```

Athlete's Height distribution over the years.

```python
sns.histplot(main_df,x="Height",bins=20,kde=True,color='#f54278')
plt.ylabel("No. of Athletes")
plt.xlabel("Height (cm)")
plt.title("Athlete's Distribution by Height (cm)")
plt.show()
```

Average height of athlete's with repect to their sport.

```python
avg_height_sport=main_df.groupby("Sport")["Height"].mean()
avg_height_sport.head(10)
```

Event with the highest average height in Olympics.

```python
max=avg_height_sport.max()
avg_height_sport[avg_height_sport==max].to_frame()
```

Tallest athlete to participate in olympics from 1896 to 2016.

```python
tallest_athlete=main_df[main_df['Height']==main_df['Height'].max()]
print('Tallest Athlete to participate in Olympics :-')
print(tallest_athlete[['ID','Name','Height','Sport']])
```

Athlete's Weight distribution over the years.

```python
sns.histplot(main_df,x="Weight",bins=20,kde=True,color='#42f5b9')
plt.ylabel("No. of Athlete's")
plt.xlabel("Weight (kg)")
plt.title("Athlete's Distribution by Weight (kg)")
plt.show()
```

Average weight of Athlete's based on their gender and event.                                                                     
Here we have taken the example of finding average weight of female who has event as Football.

```python
weight_sport_gen=main_df.groupby(['Sport','Sex'])['Weight'].mean()
weight_sport_gen.head(10)
print("Average weight of Female Athlete in Football is",weight_sport_gen['Football']['F'])
```

Heaviest Athlete participated in Olympics.

```python
athlete_weight=main_df[main_df["Weight"]==main_df['Weight'].max()]
print('Heaviest Athlete to participate in Olympics')
print(athlete_weight[['ID','Name','Height','Sport']])
```

Number of Athlete's participate along the year's

```python
paticipation_year.plot(kind='bar',color='#089c59')
plt.ylabel=("No. of Athlete's")
plt.title("Number of Athlete's Participating each year" )
plt.show()
```

Distribution of athlete's age in Summer and Winter Olympics

```python
sns.boxplot(main_df,x='Season',y='Age')
plt.title('Distribution of Ages by Season')
plt.xlabel=('Season- Summer and Winter')
plt.ylabel=("Age of Athlete's")
plt.show()
```

# Olympic Medal Analysis

Number of medals given over the years.

```python
print('Total Number of Medals given over years')
main_df.groupby(['Medal'])['Medal'].count()
```

Number of medals given in each year.

```python
sns.countplot(main_df,x='Year',hue='Medal')
plt.title('Medal Distributed Over the Years')
plt.xticks(rotation=75)
plt.ylabel("No. of Medals")
plt.show()
```

Country that received Highest number of gold medal over the years.

```python
country_medal=main_df[main_df['Medal']=='Gold'].groupby('Country')['Medal'].count()
gold_max=country_medal.max()
print('Country with Highest gold medals')
country_medal[country_medal==gold_max]
```

Distribution of Medal's by Age of athletes. 

```python
sns.violinplot(main_df,x='Medal',y='Age',palette='Set1')
plt.title('Distribution of Age by Medal')
plt.show()
```

Athlete Height and Weight factor in winning medals

```python
sns.scatterplot(main_df,x='Height',y='Weight',hue='Medal',palette='Set2')
plt.title('Athlete Height and Weight factor in winning Medal')
plt.legend(title='Medal')
plt.show()
```

Medals received by the Athlete's of various country alonf the years.

```python
medal_year_country=main_df.pivot_table(index='NOC',columns='Year',values="Medal", aggfunc='count')
sns.heatmap(medal_year_country,cmap='inferno',linewidths=0.5)
plt.xticks(rotation=50)
plt.show()
```
